```python
import os
from llama_index.core import Settings, VectorStoreIndex, StorageContext, load_index_from_storage
from llama_index.embeddings.gemini import GeminiEmbedding
from llama_index.llms.gemini import Gemini
from google.colab import userdata # To get the API key from Colab secrets

# --- Prerequisites ---
# 1. Ensure the 'documents' variable exists from the previous cell (Data Loading).
# 2. Ensure your GOOGLE_API_KEY is set up in Colab Secrets.
# --- End Prerequisites ---

# --- Configuration ---
# Retrieve the API key from Colab secrets
try:
    GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')
    if not GOOGLE_API_KEY:
        raise ValueError("GOOGLE_API_KEY not found in Colab Secrets.")
    os.environ['GOOGLE_API_KEY'] = GOOGLE_API_KEY # Optional: Set it as an env var for some libraries
    print("Google API Key loaded.")
except Exception as e:
    print(f"Error loading Google API Key from Colab Secrets: {e}")
    print("Please ensure the secret 'GOOGLE_API_KEY' is created and notebook access is enabled.")
    # Stop execution if the key is missing
    raise SystemExit("API Key configuration failed.")

# Define the LLM and Embedding models
# Check Google AI Studio for the latest available model names
LLM_MODEL_NAME = "models/gemini-2.0-flash"
EMBEDDING_MODEL_NAME = "models/text-embedding-004" # Or other suitable embedding model

# Configure LlamaIndex settings globally
print(f"Configuring LLM: {LLM_MODEL_NAME}")
Settings.llm = Gemini(model_name=LLM_MODEL_NAME, api_key=GOOGLE_API_KEY)

print(f"Configuring Embedding Model: {EMBEDDING_MODEL_NAME}")
Settings.embed_model = GeminiEmbedding(model_name=EMBEDDING_MODEL_NAME)

# Optional: Configure chunk size (default is 1024)
# Settings.chunk_size = 512
# Settings.chunk_overlap = 20
# print(f"Using chunk size: {Settings.chunk_size}, overlap: {Settings.chunk_overlap}")

# --- Indexing ---
PERSIST_DIR = "./storage" # Directory to save the index

# Check if the 'documents' variable exists before proceeding
if 'documents' not in locals() or not documents:
    print("Error: 'documents' variable not found or is empty.")
    print("Please run the Data Loading cell successfully first.")
    # Stop execution if documents aren't loaded
    raise SystemExit("Document loading failed or skipped.")
else:
    # Check if storage already exists to avoid re-indexing
    if not os.path.exists(PERSIST_DIR):
        print(f"Creating new index from {len(documents)} documents...")
        # Create the VectorStoreIndex: This performs chunking and embedding
        index = VectorStoreIndex.from_documents(
            documents,
            show_progress=True # Shows progress bar during indexing
        )
        print(f"Index created successfully.")
        # Persist the index to disk
        print(f"Persisting index to '{PERSIST_DIR}'...")
        index.storage_context.persist(persist_dir=PERSIST_DIR)
        print("Index persisted.")
    else:
        print(f"Loading existing index from '{PERSIST_DIR}'...")
        # Load the existing index
        storage_context = StorageContext.from_defaults(persist_dir=PERSIST_DIR)
        index = load_index_from_storage(storage_context)
        print("Index loaded successfully.")

# Now the 'index' variable holds your queryable index.
```

```python
# --- Prerequisites ---
# 1. Ensure the 'index' variable exists from the previous cell (Indexing).
# --- End Prerequisites ---

# Check if the 'index' variable exists
if 'index' not in locals():
    print("Error: 'index' variable not found.")
    print("Please run the Indexing cell successfully first.")
    # Stop execution if index isn't available
    raise SystemExit("Index creation/loading failed or skipped.")
else:
    print("Creating query engine...")
    # Create a query engine from the index
    # This uses the LLM and Embeddings configured in Settings
    query_engine = index.as_query_engine(
        # Optional: Enable streaming for token-by-token output
        # streaming=True,
        # Optional: Adjust the number of retrieved documents (default is 2)
        # similarity_top_k=3
    )
    print("Query engine created.")

    # --- Perform a Query ---
    print("\n--- Running Sample Query ---")
    # Replace this with a question relevant to your loaded documents
    sample_query = "What is the main topic discussed in the documents? Be detailed"
    print(f"Query: {sample_query}")

    # Execute the query
    response = query_engine.query(sample_query)

    # Print the response
    # If streaming=True, use response.print_response_stream()
    print("\nResponse:")
    print(response)

    # Optional: Inspect the source nodes used for the answer
    # print("\nSource Nodes:")
    # if hasattr(response, 'source_nodes'):
    #     for i, node in enumerate(response.source_nodes):
    #         print(f"--- Node {i+1} (Score: {node.score:.4f}) ---")
    #         print(node.text[:300] + "...") # Print start of the source text
    # else:
    #     print("No source nodes available in the response object.")
    print("--- End Sample Query ---")
```

```python
# --- Prerequisites ---
# 1. Ensure the 'index' variable exists from the previous cell (Indexing).
# --- End Prerequisites ---

# Check if the 'index' variable exists
if 'index' not in locals():
    print("Error: 'index' variable not found.")
    print("Please run the Indexing cell successfully first.")
    # Stop execution if index isn't available
    raise SystemExit("Index creation/loading failed or skipped.")
else:
    print("Creating chat engine...")
    # Create a chat engine (maintains conversation history)
    # 'context' mode uses the retriever to fetch context before generating response
    # 'verbose=True' shows intermediate steps (retrieval, LLM prompt)
    chat_engine = index.as_chat_engine(
        chat_mode='context',
        verbose=True
    )
    print("Chat engine created. Type 'quit' or 'exit' to end the chat.")

    # --- Start Interactive Chat Loop ---
    while True:
        try:
            user_input = input("You: ")
            if user_input.lower() in ["quit", "exit"]:
                print("Bot: Goodbye!")
                break
            # Send user message to chat engine and get response
            response = chat_engine.chat(user_input)
            print(f"Bot: {response}")
        except EOFError:
            # Handle Ctrl+D or similar interruption in Colab input
            print("\nBot: Chat interrupted. Goodbye!")
            break
        except Exception as e:
            print(f"An error occurred: {e}")
            # Optionally break the loop on error, or continue
            # break

    # Optional: Reset chat history if you want to start a new conversation
    # chat_engine.reset()
    # print("Chat history reset.")
```
